```python
# Different types of plots ie scatter, bar, hist etc
fav_foods = {"Fruits": 4,
             "Nuts": 10,
             "Vegetables":8}
fig, ax = plt.subplots()
ax.bar(fav_foods.keys(), fav_foods.values())

ax.set(title="Type Of Foods & Prices",
       xlabel="Foods",
       ylabel="Vegetables")

plt.show()
```

```python
# Same plot, but this time bars go horizontal
fav_foods = {"Fruits": 4,
             "Nuts": 10,
             "Vegetables":8}
fig, ax = plt.subplots()
ax.barh(list(fav_foods.keys()), list(fav_foods.values()))

ax.set(title="Type Of Foods & Prices",
       xlabel="Foods",
       ylabel="Vegetables")

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python

# Importing heart-disease.csv and save it to the variable "heart_disease"
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
# View the first 10 rows of the heart_disease DataFrame
heart_disease.head(10)
```

```python
heart_disease["age"].plot.hist(bins=50)
plt.show()
```

```python
heart_disease.plot.hist()
plt.show()
```

```python
# Called plot.hist() on the heart_disease DataFrame and toggled the
# "subplots" parameter to True and gave figsize as 10,30
heart_disease.plot.hist(figsize=(10,20), subplots=True)
plt.show()
```

```python
# Creating DataFrame with patients over 50 years old
over_50 = heart_disease[heart_disease["age"] > 50]
```

```python
over_50.head(5)
```

```python
# Creating the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x = over_50["age"], y= over_50["chol"], c= over_50["target"])

# Customizing the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel= "Age",
       ylabel= "Cholesterol")
ax.legend(*scatter.legend_elements(), title="Target")

# Adding a mean line
ax.axhline(over_50["chol"].mean(), linestyle="--")

# Adding a title
fig.suptitle("Heart Disease Analysis", fontsize= 18, fontweight="bold")


plt.show()
       
```

```python

# Creating the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data and added an cmap for color dotting
scatter = ax.scatter(x = over_50["age"], y= over_50["chol"], c= over_50["target"], cmap="winter")

# Customizing the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel= "Age",
       ylabel= "Cholesterol")
ax.legend(*scatter.legend_elements(), title="Target")

# Adding a mean line
ax.axhline(over_50["chol"].mean(), linestyle="--", color="red")

# Adding a title
fig.suptitle("Heart Disease Analysis", fontsize= 18, fontweight="bold")

# Using seaborn-whitegrid style
plt.style.use('seaborn-v0_8-whitegrid')

plt.show()
```
